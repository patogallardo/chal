# Example: one layer of silicon in vacuum

```python
from chal import stack
import numpy as np
import matplotlib.pyplot as plt
```

One layer of silicon $\epsilon=11.68$ and thickness $d=1 mm$

```python
epsilons = [[1,0], [11.68, 0], [1,0]] # epsilons is a list of real and imaginary part of epsilon
ds=[1e-3] # one layer of thickness 1mm (in meters)
fs = np.linspace(700e9, 900e9, 1000)
```

```python
s = stack.Stack(epsilons, ds, fs)
```

```python
plt.plot(s.freqs/1e9, s.T, label='1mm sample')
plt.axhline(s.T.mean(), label='mean', color='C1')
plt.legend(loc='upper right')
plt.title("Transmittance for a 1 mm thick silicon piece")
plt.xlabel('frequency [GHz]')
plt.ylabel('Transmittance [-]')
plt.savefig('Example1_Transmittance_1mmThickSilicon.pdf')
```
